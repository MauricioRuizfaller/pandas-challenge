```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
schools = pd.read_csv('schools_complete.csv')
students = pd.read_csv('students_complete.csv')
```

# District Summary

```python
school_total = schools['school_name'].nunique()
school_total
```

```python
students_total = students['Student ID'].count()
students_total
```

```python
math_avg = students['math_score'].mean()
math_avg
```

```python
reading_avg = students['reading_score'].mean()
reading_avg
```

```python
passing_math_count = students[(students["math_score"] >= 60)].count()["student_name"]
passing_math_percentage = (passing_math_count / students_total) * 100
passing_math_percentage
```

```python
passing_reading_count = students[(students["reading_score"] >= 60)].count()["student_name"]
passing_reading_percentage = (passing_reading_count / students_total) * 100
passing_reading_percentage
```

```python
passing_math_reading_count =  students[(students["math_score"] >= 60) & (students["reading_score"] >= 60)].count()["student_name"]
overall_passing_rate = (passing_math_reading_count / students_total) * 100
overall_passing_rate
```

# School Summary

```python
merged_df = pd.merge(students, schools, on='school_name')
merged_df
```

```python
schools
```

```python
school_type = merged_df.groupby('school_name')['type'].unique()
school_type
```

```python
student_amount = merged_df.groupby('school_name')['student_name'].size()
student_amount
```

```python
school_budget = merged_df.groupby('school_name')['budget'].count()
school_budget
```

```python
merged_df['Budget per Student'] = merged_df['budget'] / merged_df['size']
merged_df
```

```python
avg_mathscore = merged_df.groupby('school_name')['math_score'].mean()
avg_mathscore
```

```python
avg_readingscore = merged_df.groupby('school_name')['reading_score'].mean()
avg_readingscore
```

```python
math_passing = merged_df[(merged_df["math_score"] >= 60)]
passing_math_percentage = (math_passing.groupby('school_name').size() / merged_df.groupby('school_name').size()) * 100
passing_math_percentage
```

```python
reading_passing = merged_df[(merged_df["reading_score"] >= 60)]
passing_reading_percentage = (reading_passing.groupby('school_name').size() / merged_df.groupby('school_name').size()) * 100
passing_reading_percentage
```

```python
passing_math_reading = merged_df[(merged_df['math_score'] >= 70) & (merged_df['reading_score'] >= 70)]
overall_passing_percentage = (passing_math_reading.groupby('school_name').size() / merged_df.groupby('school_name').size()) * 100
overall_passing_percentage
```

```python
results = pd.DataFrame({
    'School Name': overall_passing_percentage.index,
    'Type': school_type,
    'Total students': student_amount,
    'Total school budget': school_budget,
    'Per student budget': school_budget,
    'Average math score': avg_mathscore,
    'Average reading score': avg_readingscore,
    '% passing math': passing_math_percentage,
    '% passing reading': passing_reading_percentage,
    '% overall passing': overall_passing_percentage
})

result = results.set_index('School Name')
result
```
